```python
import gdown
import zipfile

# Download the zip file from Drive
gdown.download(
    "https://drive.google.com/uc?id=14iGl-aEfr7NOw24Ac962J_IprESEYVGl",
    "dataset.zip",
    quiet=False
)

# Unzip into the current directory
with zipfile.ZipFile('dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('./')

print("✅ Dataset successfully downloaded and extracted!")
```

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image
```

Load Dataset

```python
df = pd.read_csv("dataset/Training_set.csv")
df.head()
```

Show Dataset Info

```python
print("Total images:", len(df))
print("Unique labels:", df['label'].nunique())
df['label'].value_counts()
```

Plot Class Distribution

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Create the directory if it doesn't exist
if not os.path.exists('screenshots'):
    os.makedirs('screenshots')

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='label', order=df['label'].value_counts().index)
plt.title("Butterfly Species Distribution")
plt.xticks(rotation=45, ha='right') # Rotate and align x-axis labels
plt.tight_layout()
plt.savefig("screenshots/data_distribution.png")
plt.show()
```

Preview Few Images

```python
from IPython.display import display

img_dir = "dataset/train"
sample_df = df.sample(5)

for i, row in sample_df.iterrows():
    img_path = os.path.join(img_dir, row['filename'])
    img = Image.open(img_path)
    print(f"Label: {row['label']}")
    display(img)
```
